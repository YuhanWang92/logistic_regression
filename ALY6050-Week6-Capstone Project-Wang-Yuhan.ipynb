```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import log
```

```python
hon = pd.read_csv('W6_Data.csv')
price = hon['Close Price']
date = hon['Date']
```

```python
import statsmodels.api as sm
#decomposition
#smoothing parameter = 1600 for quarterly time series
cycle, trend = sm.tsa.filters.hpfilter(price,1600)
```

# Original Stock Price 

```python
#ORIGINAL TIME SERIES PLOT
plt.subplot(3, 1, 1)
#plt.rcParams["figure.figsize"] = fig_size
plt.plot_date(x=date, y=price, fmt="r-") #A plot with data that contains dates.
plt.title("Stock:HON - Hodrick Prescott Filter Analysis")
plt.ylabel("Price")
plt.xlabel("Date")
plt.grid(True)
plt.show()
```

# Trend & Cyclic Components

```python
#plot the decomposition results
#TREND
plt.subplot(3, 1, 2)
#plt.rcParams["figure.figsize"] = fig_size
plt.plot_date(x=date, y=trend, fmt="r-")
plt.title("Stock:HON - Trend Component")
plt.ylabel("Trend")
plt.xlabel("Date")
plt.grid(True)
plt.show()

#CYCLE
plt.subplot(3, 1, 3)
#plt.rcParams["figure.figsize"] = fig_size
plt.plot_date(x=date, y=cycle, fmt="r-")
plt.title("Stock:HON - Cycle Component")
plt.ylabel("Cycle")
plt.xlabel("Date")
plt.grid(True)
plt.show()
```

# Original & Trend

```python
#Plot the line plots of the original time series along with its trend component on the same chart
ori_cycle, ori_trend = sm.tsa.filters.hpfilter(price,1600)#the decomposition with original price
price_trend = pd.DataFrame(price)
price_trend['trend'] = trend #append the trend component with orginal price
price_trend.plot(figsize=(10,6))
plt.title('Original & Trend')
plt.ylabel("Price")
plt.show()
```
